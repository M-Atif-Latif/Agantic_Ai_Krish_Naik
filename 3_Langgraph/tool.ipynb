# Langchain tools

```python
from langchain_community.tools import WikipediaQueryRun
```

```python
from langchain_community.utilities import WikipediaAPIWrapper
```

```python
api_wrapper=WikipediaAPIWrapper(top_k_results=5 , doc_content_chars_max=1000)
```

```python
wiki_tool=WikipediaQueryRun(api_wrapper=api_wrapper)
```

```python
wiki_tool.name
```

```python
wiki_tool.description
```

```python
wiki_tool.run("Pakistan")
```

```python
# Query Wikipedia for information about Pakistan
result = wiki_tool.run("Pakistan")
print("Wikipedia search results for Pakistan:")
print(result)
```
